```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[1],[0],[0],[0]], "float32")
```

```python
model = Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])
```

```python
model.fit(training_data, target_data, epochs=200)
```

```python
scores = model.evaluate(training_data, target_data)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
print("AND Gate ANN Predicted Values:")
print(model.predict(training_data).round())
```

```python
X1 = np.array([[0,0]])
X2 = np.array([[0,1]])
X3 = np.array([[1,0]])
X4 = np.array([[1,1]])
print("Truth Table\n\n  A B\t\tA NOR B")
print(X1,"\t", model.predict_classes(X1))
print(X2,"\t", model.predict_classes(X2))
print(X3,"\t", model.predict_classes(X3))
print(X4,"\t", model.predict_classes(X4))
print("\n______________________\n")
```
