```
#Separating feature columns from label
feature_cols = ['radius_mean', 'texture_mean', 'perimeter_mean', 
                'area_mean','smoothness_mean','compactness_mean','concavity_mean',
                'concave points_mean','symmetry_mean','fractal_dimension_mean','radius_se',
                'texture_se','perimeter_se','area_se','smoothness_se','compactness_se',
                'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 
                'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 
                'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst',
                'fractal_dimension_worst']
X = cancer[feature_cols]
y = cancer.diagnosis
```

```
#Splitting data for testing and training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```
#KNN model for training data and accuracy on testing data for neighbors ranging from 2-10
for i in range(2,10):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn = knn.fit(X_train,y_train)
  y_test_pred = knn.predict(X_test)
  y_train_pred = knn.predict(X_train)
  print("Accuracy for ", i, "neighbors: ",knn.score(X_test,y_test))

print("\nWe get highest accuracy for k = 7 neighbors at 93.56%")
```
